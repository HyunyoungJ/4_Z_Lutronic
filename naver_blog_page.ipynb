```python
query = input("query : ")
pages = int(input("pages_*7 : "))


list_page = []
i=0
for i in range(1,pages + 1):
    list_page.append(i)
    
list_page

#입력 데이터(query), 돌아다닐 블로그 개수 : pages_*7
```

```python
# url 가공
def get_final_url(url_1):
        
    try :
        html_result = requests.get(url_1)
        soup_temp = BeautifulSoup(html_result.text,'html.parser')
        area_temp = soup_temp.find(id='mainFrame')
        url_3 = area_temp.get('src')
        url_4 = "http://blog.naver.com" + url_3
        
        return url_4
    except : 
        print("error")
        return None

 
```

```python
# soup = BeautifulSoup(url.text, 'html.parser')
time.sleep(5)
# area = soup.find("div",{"class":"total_wrap api_ani_send"}).find_all("a",{"class":"api_txt_lines total_tit"})
id_num = 1
url_list=[]



title = []


for item in list_page:
    num_page = item
    url = f'https://section.blog.naver.com/Search/Post.nhn?pageNo={num_page}&rangeType=ALL&orderBy=recentdate&keyword={query}'

    print(url)
    driver.get(url)
    time.sleep(2)
    try : 
    #     area = soup.find("li",{"id":f"sp_blog_{id_num}"}).find_all("a",{"class":"api_txt_lines total_tit"})
        area = driver.find_elements_by_class_name('desc_inner')
        for i in range(0,7):
            url1=area[i].get_attribute('href')
            title1 = area[i].find_element_by_class_name('title').text
            print(url1)
            print(title1)
            a =get_final_url(url1)
            url_list.append(a)
            title.append(title1)

    except:
        pass
print(url_list)

#검색 옵션 변경시 url 에서 range type, orderBy부분을 바꿔주시면 됩니다.
df1 = pd.DataFrame(title) #타이틀 누락 문제로 추가 파일 생성합니다., 추후 merge시키거나 복사 붙여넣기 하면 될 듯합니다.
df1.to_csv(f'\data_blog_{query}.csv', sep=',',na_rep='NaN', encoding='utf-8-sig')
```
