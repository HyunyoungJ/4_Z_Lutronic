```python
url = 'https://search.naver.com/search.naver'
# https://search.naver.com/search.naver?where=post&smtab_jum&query=%EC%8B%9C%EC%88%A0
hrd = { 'User-Agent' : 'Mozilla/5.0', 'referer' : 'https://naver.com'}
param = {
    'where' : 'post',
    'query' : '체형관리',
    'date_from' : '20180401',
    'data_to' : '20200813',
    'date_option' : '8'
} # query : 검색어, date_from : 어느시점부터 검색, date_to : 현재까지(작동안함)


#date_to는 작동을 안하는듯하다
response = requests.get(url, params=param, headers=hrd)
print(response) # 요청결과 출력
print(response.url)# 전달한 매개변수가 포함된 url 형태 출력

print(response.status_code) #요청코드 출력
# print(response.text) #해당 url로의 html코드 출력


# &query=시술&sm=tab_view.cafe
```

```python
driver.get(response.url)
SCROLL_PAUSE_TIME = 2

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

target_trick = time.time() + 30

while time.time()<target_trick:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        try:
            driver.find_element_by_css_selector(".mye4qd").click()
        except:

            break
    last_height = new_height
```

```python
# url 가공
def get_final_url(url_1):
        
    try :
        html_result = requests.get(url_1)
        soup_temp = BeautifulSoup(html_result.text,'html.parser')
        area_temp = soup_temp.find(id='mainFrame')
        url_3 = area_temp.get('src')
        url_4 = "http://blog.naver.com" + url_3
        
        return url_4
    except : 
        print("error")
        return None

 
```
